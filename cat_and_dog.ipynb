```
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image 
import cv2
from tqdm import tqdm

DATADIR = '/content/gdrive/My Drive/data/dogs'
dogs=[]
for j in os.listdir(DATADIR):
  img=cv2.imread(os.path.join(DATADIR,j))
  img=cv2.resize(img,(224,224))
  
  
  dogs.append(img)




```

```
DATADIR = '/content/gdrive/My Drive/data/cats'
cats=[]
for j in os.listdir(DATADIR):
  img=cv2.imread(os.path.join(DATADIR,j))
  img=cv2.resize(img,(224,224))
  
  
  cats.append(img)



  
 
```

```
data=dogs+cats
data=np.array(data)
```

```
import sys
from matplotlib import pyplot
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from keras.optimizers import SGD
from keras.preprocessing.image import ImageDataGenerator
from keras.callbacks import EarlyStopping, ReduceLROnPlateau
```

```
y=[]
for i in range(0,len(cats)):
  y.append(0)
for i in range(0,len(dogs)):
  y.append(1)
```

```
from tensorflow.keras.applications.vgg19 import VGG19
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.vgg19 import preprocess_input
from tensorflow.keras.models import Model
from sklearn.model_selection import train_test_split
model=VGG19(weights="imagenet",include_top=False,pooling="avg")
training_features=model.predict(data)

x_train,x_test,y_train,y_test=train_test_split(training_features, y, test_size=0.25, random_state=2)

from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
naive_bayes=GaussianNB()
naive_bayes.fit(x_train,y_train)
y_pred=naive_bayes.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)*100
print(accuracy)
```
